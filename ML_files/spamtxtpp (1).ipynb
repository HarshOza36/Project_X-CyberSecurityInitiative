```python
corpus = []
ps = PorterStemmer()
```

```python
print (df2['text'][0])
print (df2['text'][1])
```

```python
for i in range (0,5814):
    m = df2['text'][i]
    m = re.sub('\b[\w\-.]+?@\w+?\.\w{2,4}\b', 'emailadd', df2['text'][i])
    m = re.sub('(http[s]?\S+)|(\w+\.[A-Za-z]{2,4}\S*)', 'urladd', df2['text'][i])
    m = re.sub('£|\$', 'moneysymbols', df2['text'][i])
    m = re.sub('\b(\+\d{1,2}\s)?\d?[\-(.]?\d{3}\)?[\s.-]?\d{3}[\s.-]?\d{4}\b', 'phoneno', df2['text'][i])
    m = re.sub('\d+(\.\d+)?', 'numbers', df2['text'][i])
    m = re.sub('[^\w\d\s]', ' ', df2['text'][i])    
#    if i<4:
#        print("MESSAGE ", i)
#    
#    if i<4:
#        print("\n After Regular Expression - Message ", i, " : ", m)   
    m = m.lower().split()    
#    if i<4:
#        print("\n Lower case Message ", i, " : ", m)
    
#    m = m.split()    
#    if i<4:
#        print("\n After Splitting - Message ", i, " : ", m)
    
    m = [ps.stem(word) for word in m if not word in set(stopwords.words('english'))]
#    if i<4:
#        print("\n After Stemming - Message ", i, " : ", m)
#    
    m = ' '.join(m)
#    if i<4:
#        print("\n Final Prepared - Message ", i, " : ", m, "\n\n")
    
    corpus.append(m)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cv = CountVectorizer()
x = cv.fit_transform(corpus).toarray()
x
```

```python
y = df2['type']
print (y.value_counts())

print(y[0])
print(y[1])
```

```python
le = LabelEncoder()
y = le.fit_transform(y)

print(y[0])
print(y[1])
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y,test_size= 0.15, random_state = 0)
xtrain
xtest.shape
```

```python
bayes_classifier = GaussianNB()
bayes_classifier.fit(xtrain, ytrain)
print(type(xtrain))
```

```python
# Predicting
y_pred = bayes_classifier.predict(xtest)
y_pred
```

```python
# Evaluating
cm = confusion_matrix(ytest, y_pred)
print(cm)
print ("Accuracy : %0.5f \n\n" % accuracy_score(ytest,y_pred))
```

```python
dt = DecisionTreeClassifier(random_state=50)
dt.fit(xtrain, ytrain)
```

```python
# Predicting
y_pred_dt = dt.predict(xtest)
```

```python
# Evaluating
cm = confusion_matrix(ytest, y_pred_dt)

print(cm)
```

```python
from sklearn.metrics import classification_report
print ("Accuracy : %0.5f \n\n" % accuracy_score(ytest, dt.predict(xtest)))
print (classification_report(ytest, dt.predict(xtest)))
```
