```python
corpus = []
ps = PorterStemmer()
```

```python
for i in range (0,len(df)):
    m = df['text'][i]
    m = re.sub('\b[\w\-.]+?@\w+?\.\w{2,4}\b', 'emailadd', df['text'][i])
    m = re.sub('(http[s]?\S+)|(\w+\.[A-Za-z]{2,4}\S*)', 'urladd', df['text'][i])
    m = re.sub('£|\$', 'moneysymbols', df['text'][i])
    m = re.sub('\b(\+\d{1,2}\s)?\d?[\-(.]?\d{3}\)?[\s.-]?\d{3}[\s.-]?\d{4}\b', 'phoneno', df['text'][i])
    m = re.sub('\d+(\.\d+)?', 'numbers', df['text'][i])
    m = re.sub('[^\w\d\s]', ' ', df['text'][i])    
    m = m.lower()    
    m = m.split()    
    m = [ps.stem(word) for word in m if not word in set(stopwords.words('english'))]
    m = ' '.join(m)
    corpus.append(m)
```

```python
corpus
len(df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
cv = CountVectorizer()
x = cv.fit_transform(corpus).toarray()
```

```python
y = df['spam']
print (y.value_counts())

#print(y[0])
#print(y[1])
#print(y)
```

```python
'''df['spam'].replace('<','1')
df['spam'].replace('>','1')
df['spam'].replace('NaN','1')'''
```

```python
le = LabelEncoder()
y = le.fit_transform(y)

print(y[0])
print(y[5700])
y
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size= 0.20, random_state = 0)
```

```python
bayes_classifier = GaussianNB()
bayes_classifier.fit(xtrain, ytrain)
```

```python
# Predicting
y_pred = bayes_classifier.predict(xtest)
print(y_pred)
```

```python
# Evaluating
cm = confusion_matrix(ytest, y_pred)
print(cm)
print ("Accuracy : %0.5f \n\n" % accuracy_score(ytest, bayes_classifier.predict(xtest)))
```
